```python
from tkinter import *
from tkinter import messagebox
users={}
class mainPage():
    def __init__(self,parent):
        self.parent=parent
        self.Label1=Label(self.parent,text="Registration Form",font=("Arial Bold",20)).grid(row=0,column=1)
        self.Label2=Label(self.parent,text="Name").grid(row=1)
        self.Label3=Label(self.parent,text=" UserName").grid(row=2)
        self.Label4=Label(self.parent,text="Phone").grid(row=3)
        self.Label5=Label(self.parent,text="Email").grid(row=4)
        self.Label6=Label(self.parent,text="Password").grid(row=5)

        self.ename=Entry(self.parent)
        self.ename.grid(row=1,column=1)
        self.euname=Entry(self.parent)
        self.euname.grid(row=2,column=1)
        self.ephone=Entry(self.parent)
        self.ephone.grid(row=3,column=1)
        self.eemail=Entry(self.parent)
        self.eemail.grid(row=4,column=1)
        self.epassword=Entry(self.parent,show="*")
        self.epassword.grid(row=5,column=1)
        self.Button1=Button(self.parent, text='Reset', command=self.resetForm).grid(row=6, column=0, sticky=W, pady=4)
        self.Button2=Button(self.parent, text='Submit', command=self.registerUser).grid(row=6, column=1, sticky=W, pady=4)
    def resetForm(self):
        self.ename.delete(0, 'end')
        self.euname.delete(0,'end')
        self.eemail.delete(0,'end')
        self.ephone.delete(0,'end')
        self.epassword.delete(0,'end')  
    def loginUser(self):
        login=Toplevel(self.parent)
        loginPage=LoginPage(login)
    def registerUser(self):
        self.ulist=[]
        self.ulist.append(self.ename.get())
        self.ulist.append(self.euname.get())
        self.ulist.append(self.eemail.get())
        self.ulist.append(self.ephone.get())
        self.ulist.append(self.epassword.get())
        self.uname=self.ulist[1]
        if (self.uname in users):
            messagebox.showinfo("Invalid", "User Already Exists Go to Login")
            self.loginUser()
        else:
            users[self.uname]=self.ulist
            messagebox.showinfo("Successful", "Registered Proceed to Login")
            self.loginUser()


class RightPage():
    def __init__(self,master):
        self.master=master
        self.Label1=Label(self.master,text="Login Successful").grid(row=0)
        self.Label2=Label(self.master,text="Welcome To Booking.com").grid(row=1)
       
        
class UnRightPage():
    def __init__(self,master):
        self.master=master
        self.Label1=Label(self.master,text="Invalid Credentials").grid(row=0)
        self.Label2=Label(self.master,text="Please Check Your Credentials").grid(row=1)
        
    
    
class LoginPage():
    
    def __init__(self,master):
        self.master=master
        self.Label1=Label(self.master,text="Login Page").grid(row=0)
        self.Label2=Label(self.master,text="Username").grid(row=1,column=0)
        self.Label3=Label(self.master,text="Password").grid(row=2,column=0)
        self.uname=Entry(self.master)
        self.password=Entry(self.master,show="*")
        self.uname.grid(row=1,column=1)
        self.password.grid(row=2,column=1)
        self.button1=Button(self.master,text="Login",command=self.checkLogimn).grid(row=3)
    def checkLogimn(self):
        uname=self.uname.get()
        password=self.password.get()
        if ( (uname in users) and (users[uname][4]==password)):
            success=Toplevel(self.master)
            newPage=RightPage(success)
        else:
            unsuccess=Toplevel(self.master)
            newPage=UnRightPage(unsuccess)
        
root=Tk()
hello=mainPage(root)
root.mainloop()
    

```
